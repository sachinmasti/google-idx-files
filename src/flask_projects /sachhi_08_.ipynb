```python
import pandas as pd
from matplotlib import  pyplot as plt
import seaborn as sns
```

```python
a=np.arange(1,11)
b=np.arange(11,21)
c=np.arange(21,31)
```

```python
print(a)
print(b)
print(c)

abc=list([a,b,c])
print(abc)
```

```python
my_plot=plt.violinplot(abc,showmeans=True,showextrema=True)
color=['#87CEEB','green','pink']
for i,colors in zip(my_plot['bodies'],color):
    i.set_facecolor(colors)
    i.set_edgecolor('y')
plt.show()
```
