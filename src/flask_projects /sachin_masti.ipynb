```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('fmri')
```

```python
df.head()
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',style='event',data=df,markers=True)
```

```python
sns.barplot(x='region',y='signal',data=df,hue='event',palette= 'Blues_d')
sns.set(style= 'whitegrid')
plt.show()
```

```python
sns.scatterplot(x='timepoint',y='signal',data=df,hue='event')
plt.show()
```

```python
sns.displot(df['timepoint'],color='pink',bins=5)
plt.show()
```

```python
sns.jointplot(x='timepoint',y='event',data=df,color='pink')
plt.show
```
